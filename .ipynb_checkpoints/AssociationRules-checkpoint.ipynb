### 数据处理，从Location 1属性中提取能用的Location属性值

````python
def extractAddress (x) :
    x = str(x)
    resGroup = re.search(r'address": "(.*?)"',x)
    if resGroup !=None:
        x = resGroup.group().replace('"',"").replace("address: ","")
    else:
        x = None
    return(x)
# 提取address信息并替换 Location 1 属性的值
Crime12DF['Location 1'] = Crime12DF['Location 1'].apply(extractAddress)
Crime14DF['Location 1'] = Crime14DF['Location 1'].apply(extractAddress)
````

### 提取数据集中可分析属性

````python
## 创建需要分析的数据列表 
SubDF = CrimeDF[['location', 'areaId', 'beat', 'incidentTypeId', 'handingTime']]
#SubDF = SubDF.sample(n=200000,random_state=12315,axis=0)
````

### 对数据进行处理转换成可以进行关联规则挖掘的形式，对于数值型数据进行分层

````python
def normalizef(x):
    x['location'] = "Location_"+x['location'].strip()
    x['beat'] = 'Beat_'+x['beat'].strip()
    x['incidentTypeId'] = 'IncTypeId_'+x['incidentTypeId'].strip()
    x['areaId'] = 'AreaId_'+str(x['areaId'])
#     x['priority'] = 'Priority' + str(x['priority'])
    s = x['handingTime']
    if s > 60*24*3:
        s = 'HandingTime'+'3d'
    elif (s > 60*24*2) and (s <= 60*24*3):
        s = 'HandingTime_'+'2d_3d'
    elif s > 60*24 and s <= 60*24*2:
        s = 'HandingTime_'+'1d_2d'
    elif s >60*20 and s<=60*24:
        s = 'HandingTime_'+'20h_24h'
    elif s >60*16 and s<=60*20:
        s = 'HandingTime_'+'16h_20h'
    elif s >60*12 and s<=60*16:
        s = 'HandingTime_'+'12h_16h'
    elif s >60*8 and s<=60*12:
        s = 'HandingTime_'+'8h_12h'
    elif s >60*4 and s<=60*8:
        s = 'HandingTime_'+'4h_8h'
    elif s >60*1 and s<=60*4:
        s = 'HandingTime_'+'1h_4h'
    elif s<=60:
        s = 'HandingTime_'+'1h'
    x['handingTime'] = s
    return x
SubDF = SubDF.apply(normalizef, axis=1)
````

````python
SubDF
````

### 对全部不重复属性值进行One-hot编码，并生成编码和解码字典，方便规则的转换。然后将所有元组转换成向量表示

````python
strSet = SubDF['location'].unique().tolist()+SubDF['areaId'].unique().tolist()+SubDF['beat'].unique().tolist()+SubDF['incidentTypeId'].unique().tolist()+SubDF['handingTime'].unique().tolist()
#+SubDF['priority'].unique().tolist()
````

````python
import orangecontrib.associate.fpgrowth as oaf  #进行关联规则分析的包 

strEncode = dict(zip(strSet,range(len(strSet)))) #编码字典
strDecode = dict(zip(strEncode.values(), strEncode.keys()))  #解码字典

def toNum(x):
    x['location'] = strEncode[x['location']]
    x['beat'] = strEncode[x['beat']]
    x['incidentTypeId'] = strEncode[x['incidentTypeId']]
    x['areaId'] = strEncode[x['areaId']]
#     x['priority'] =strEncode[x['priority']]
    x['handingTime'] = strEncode[x['handingTime']]
    return x

# 将全部元组转化成向量表示
AnalysisDF = SubDF.apply(toNum, axis=1)
````

### 在支持度为0.0025和置信度为0.5的条件下，进行关联规则的挖掘

````python
itemsets = dict(oaf.frequent_itemsets(AnalysisDF.values.tolist(),  .0025))
rules = oaf.association_rules(itemsets, .65)   #设置置信度
rules = list(rules)
````

````python
print(len(rules))
````

````python
# 查看频繁项集
itemsets
````

### 打印挖掘出的规则, 并使用Lift、Kulc和IR指标，评价关联规则

````python
 def dealRules(rules,strDecode):
    returnRules = []
    for i in rules:
        temList = []
        pb = i[3]/i[6]
        pa = i[4]
        pab = i[3]*pa
        temStr = '( ';
        for j in i[0]:   #处理第一个frozenset
            temStr = temStr+strDecode[j]+', '
        temStr = temStr[:-2]+' )'
        temStr = temStr + ' ==> ('
        for j in i[1]:
            temStr = temStr+strDecode[j] + ', '
        temStr = temStr[:-2]+' )'
        temList.append(temStr); temList.append(i[2]); temList.append(i[3]); temList.append(i[6]); 
#         temList.append(0.5*(i[3]+i[3]/i[5]))
        temList.append(0.5*(pab/pa+pab/pb)) 
        temList.append(abs(pa-pb)/(pa+pb-pab))
        temList.append(pa)
        temList.append(pb)
        temList.append(pab)
        returnRules.append(temList)
    return pd.DataFrame(returnRules,columns=('规则','支持度','置信度','Lift','Kulc','IR','SupA','SupB','SupAB'))

#printRules = dealRules(rules,strDecode)
````

````python
result = list(oaf.rules_stats(rules, itemsets, len(AnalysisDF)))
````

````python
StateDF = dealRules(result,strDecode)
StateDF
````

````python
StateDF.sort_values(by='Lift' , ascending=False)
````

### 根据结果可以看到，对于第一条规则，地点EDGEWATER DR通常与31X相关性非常高，这可以从其提升度和置信度可以看出（0.997,74.5706），同时观察其Kulc和IR,发现其关系非常密切且不平衡，即地点为'EDGEWATER DR'时，beatid和可能是‘31X’, 但是当beatid是‘31X’时，地点却不太可能是‘EDGEWATER DR’。这一点可以在其支持度上得到印证，B的支持度0.01337远大于AB同时出现的支持度0.00349和A的支持度0.0035。
### 同样的道理可以解释其他规则
### 在其中可以找到一些很有意义的规则，例如规则( IncTypeId_933R, AreaId_P2 ) ==> (HandingTime_1h )说明，当案件类型是933R，且地点是P2区时，发生的案件处理时间通常少于一小时。
